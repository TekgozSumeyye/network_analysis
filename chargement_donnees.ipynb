Récupération, nettoyage et sauvegarde des données en fichier CSV pour facilement réutiliser

```python
import pickle
import pandas as pd
import glob
import os
```

```python
df= pd.DataFrame()
```

```python
dir = "C://Users//tekgo//Documents//MASTER//SEMESTRE 3//NETWORK//"

for file in os.listdir(dir):
    if file.endswith(".pickle"):
        with open(dir+file, 'rb') as f:
            data = pickle.load(f)
            df2 = pd.DataFrame(data)
            df=pd.concat([df,df2])
```

```python
print(df)
```

```python
df_select = df.loc[:, ['dcterms:identifier{Literal}','dcterms:title{Literal}','dcterms:abstract{Literal}@fr','cito:cites{URIRef}[0]','cito:cites{URIRef}[1]','cito:cites{URIRef}[2]','cito:cites{URIRef}[3]','cito:cites{URIRef}[4]','cito:cites{URIRef}[5]','cito:cites{URIRef}[6]','cito:cites{URIRef}[7]','cito:cites{URIRef}[8]','cito:cites{URIRef}[9]','marcrel:aut{URIRef}[0]','marcrel:aut{URIRef}[1]','marcrel:aut{URIRef}[2]', 'marcrel:aut{URIRef}[3]', 'marcrel:aut{URIRef}[4]', 'marcrel:aut{URIRef}[5]', 'marcrel:aut{URIRef}[6]', 'marcrel:aut{URIRef}[7]', 'marcrel:aut{URIRef}[8]', 'marcrel:aut{URIRef}[9]']]
print(df_select)
```
